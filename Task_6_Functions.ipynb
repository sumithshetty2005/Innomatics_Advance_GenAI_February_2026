### 1. Smart Parking Lot Management System

```python
def manage_parking(capacity, logs):
    current_count = 0
    for log in logs:
        if log == "IN":
            current_count += 1
        elif log == "OUT":
            current_count -= 1
    
    status = "Available" if current_count < capacity else "Full/Over Capacity"
    
    print(f"Currently Parked Vehicles: {current_count}")
    print(f"Parking Status: {status}")
    if current_count > capacity:
        print("Alert: Capacity Exceeded!")
manage_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

### 2.Online Food Delivery Time Estimator


```python
def estimate_delivery(distance, traffic, weather):
    eta = distance * 5 
    if traffic == "High": eta += 15
    elif traffic == "Medium": eta += 5
    if weather == "Rainy": eta += 10
    elif weather == "Stormy": eta += 2
    print(f"Estimated Delivery Time: {eta} minutes")

estimate_delivery(8, "High", "Rainy")
```

### 3. Movie Theatre Seat Occupancy Analyzer

```python
def analyze_occupancy(total_seats, booked_list):
    booked_count = len(booked_list)
    occupancy = (booked_count / total_seats) * 100
    
    if occupancy == 100: status = "Housefull"
    elif occupancy >= 75: status = "Almost Full"
    else: status = "Seats Available"
    
    print(f"Occupancy: {occupancy}%")
    print(f"Show Status: {status}")
analyze_occupancy(200, [1] * 150)
```

### 4. Cloud Server Load Classification System

```python
def classify_server_load(readings):
    avg_load = sum(readings) / len(readings)
    
    if avg_load < 50: status = "Normal"
    elif 50 <= avg_load <= 80: status = "Warning"
    else: status = "Critical"
    
    print(f"Average CPU Load: {avg_load}%")
    print(f"Server Status: {status}")

classify_server_load([45, 60, 70, 85, 90])
```

### 5. Smart Classroom Resource Usage Monitor

```python
def monitor_resources(usage_dict):
    threshold = 8
    overused = [res for res, hours in usage_dict.items() if hours >= threshold]
    
    print(f"Overused Resources: {', '.join(overused)}")
    print(f"Energy Alert: {'Yes' if overused else 'No'}")

monitor_resources({"Projector": 6, "AC": 9, "Lights": 4})
```

### 6. Online Event Registration Capacity Controller

```python
def manage_registrations(capacity, total_registrations):
    if total_registrations > capacity:
        confirmed = capacity
        waitlisted = total_registrations - capacity
        status = "Closed"
    else:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"
        
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")

manage_registrations(100, 105)
```
