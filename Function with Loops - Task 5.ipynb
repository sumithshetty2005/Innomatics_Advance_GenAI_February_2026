### 1. Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like
    
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

analyze_engagement([500, 200, 350])
```

### 2. Healthcare – Medicine Stock Alert System

```python
def monitor_stock(medicine_name, current_stock):
    status = "Low Stock Alert" if current_stock < 10 else "Stock Sufficient"
    
    print(f"Medicine: {medicine_name}")
    print(f"Medicine Stock: {current_stock}")
    print(f"Status: {status}")

monitor_stock("Paracetamol", 6)
```

### 3. Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall(monthly_data, required_avg):
    total_rain = 0
    for rain in monthly_data:
        total_rain += rain
    
    avg_rainfall = total_rain / len(monthly_data)
    status = "Adequate Rainfall" if avg_rainfall >= required_avg else "Inadequate Rainfall"
    
    print(f"Average Rainfall: {avg_rainfall:.0f}")
    print(f"Rainfall Status: {status}")

check_rainfall([80, 65, 71], 70)
```

### 4. Social Media – Duplicate Account Detection

```python
def detect_duplicates(usernames):
    # Sets do not allow duplicates; if the length changes, a duplicate existed.
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"
    
    print(f"Duplicate Accounts Found: {result}")

detect_duplicates(["user1", "dev_king", "user1", "ace_coder"])
```

### 5. Healthcare – Appointment Eligibility Checker

```python
def check_appointment_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"
    
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")

check_appointment_eligibility(21)
```

### 6. Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(price_list):
    premium_crops = [price for price in price_list if price > 2000]
    
    print(f"Premium Crops: {premium_crops}")

filter_premium_crops([1500, 2500, 1800, 3200])
```

### 7. System Monitoring – Application Health Checker

```python
def check_app_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
        
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

check_app_health(7)
```

### 8. Banking – Daily Transaction Limit Checker

```python
def validate_transaction(amount):
    LIMIT = 50000
    status = "Approved" if amount <= LIMIT else "Rejected"
    
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

validate_transaction(60000)
```

### 9. E-Learning – Student Attendance Eligibility System

```python
def check_exam_eligibility(attendance_list):
    # Attendance_list contains 1 for present, 0 for absent
    attendance_pct = (sum(attendance_list) / len(attendance_list)) * 100
    status = "Eligible" if attendance_pct >= 75 else "Not Eligible"
    
    print(f"Attendance Percentage: {attendance_pct:.1f}")
    print(f"Exam Eligibility: {status}")

check_exam_eligibility([1, 1, 1, 1, 0]) # 4 out of 5 is 80%
```

### 10. Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    bill = 0
    
    # Calculate Slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
        
    print(f"Units Consumed: {units}")
    print(f"Total Bill: ₹{bill}")
    print(f"Usage Status: {usage}")

calculate_electricity_bill(250)
```
