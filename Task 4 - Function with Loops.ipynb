### 1. Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    while True:
        try:
            amount = int(input("Enter recharge amount (₹): "))
            if amount >= 50 and amount in valid_plans:
                print(f"Success! Recharge of ₹{amount} processed.")
                break
            else:
                print("Invalid plan. Please choose 199, 299, 399, or 599.")
        except ValueError:
            print("Error: Please enter a valid number.")
validate_recharge()
```

### 2.Inventory Reorder Alert System

```python
def inventory_alert(stock_dict):
    for item, qty in stock_dict.items():
        if qty < 15:
            print(f"Status: Reorder Alert for {item} (Stock: {qty})")
        else:
            print(f"Status: {item} Stock OK (Stock: {qty})")

inventory = {"Laptops": 20, "Mice": 12, "Keyboards": 45, "Monitors": 8}
inventory_alert(inventory)
```

### 3. Student Result Processing System

```python
def process_student_result(marks):
    total = 0
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    result = "Pass" if average >= 50 else "Fail"
    
    print(f"Average Marks: {average}")
    print(f"Final Result: {result}")

student_marks = [65, 70, 45, 80, 55]
process_student_result(student_marks)
```

### 4. Cab Fare Estimator with Retry

```python
def cab_fare_estimator():
    while True:
        km = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ").lower()
        
        fare = 50 + (km * 12)
        if peak_hour == "yes":
            fare *= 1.25  # 25% extra
            
        print(f"Total Estimated Fare: ₹{fare}")
        
        retry = input("Calculate another fare? (yes/no): ").lower()
        if retry != "yes":
            break
cab_fare_estimator()
```

### 5. Employee Attendance Checker

```python
def check_attendance_eligibility(attendance_list):
    present_days = 0
    for day in attendance_list:
        if day == "P":
            present_days += 1
            
    percentage = (present_days / len(attendance_list)) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    
    return f"Attendance: {percentage}% | Status: {status}"

attendance = ["P", "P", "A", "P", "P", "P", "A", "P"]
print(check_attendance_eligibility(attendance))
```

### 6. Password Strength Checker

```python
def password_strength_checker(password):
    if len(password) < 8:
        return "Weak: Minimum 8 characters required."
    
    has_digit = any(char.isdigit() for char in password)
    has_special = any(char in "@#$" for char in password)
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak: Must include a digit and a special character (@, #, $)."
        
pwd = input("Create a password: ")
print(password_strength_checker(pwd))
```
