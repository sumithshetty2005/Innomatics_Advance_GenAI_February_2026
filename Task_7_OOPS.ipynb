### 1: Smart Light Controller

class SmartLightController:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def set_action(self, action):
        self.status = action

    def display_status(self):
        print(f"{self.name} is {self.status}")
light = SmartLightController("Bedroom Light")
light.set_action("ON")
light.display_status()

### 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.dept}")
emp = Employee("Rahul", "EMP102", "IT")
emp.display_card()
```

### 3: Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

c = Contact("Anita", "9876543210")
c.display_contact()
```

### 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_tag(self):
        print(f"Product: {self.name}")
        # Using the Rupee symbol as requested
        print(f"Price: ₹{self.price}")

p = Product("Headphones", 2499)
p.print_tag()
```

### 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def show_rating(self):
        print(f"Movie: {self.title}")
        print(f"Rating: {self.rating} / 5")

m = Movie("Inception", 4.8)
m.show_rating()
```

### 6: Delivery Address Manager

```python
class DeliveryManager:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def print_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")
delivery = DeliveryManager("Suman", "Hyderabad")
delivery.print_details()
```
