### 1. Employee Performance Bonus Eligibility

```python
employees = {"Ravi": 92, "Ankita": 88, "Kiran": 92, "Suresh": 85}

max_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == max_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (score: {max_score})")
```

### 2. Search Query Keyword Analysis

```python
query = "Buy mobile phone buy phone online"
words = query.lower().split()
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

duplicates = {word: count for word, count in freq.items() if count > 1}

print(duplicates)
```

### 3. Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = [(index, val) for index, val in enumerate(sensor_readings) if val % 2 == 0 ]

print("Valid Sensor Readings (Hour, Value:")
print(valid_readings)
```

### 4. Email Domain Usage Analysis

```python
emails = ["ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com"]
domain_counts = {}
for email in emails:
    domain = email.split('@')[1]
    domain_counts[domain] = domain_counts.get(domain, 0) + 1

total = len(emails)
for domain, count in domain_counts.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage: .0f}%")
```

### 5. Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

avg_sales = sum(sales) / len(sales)
threshold_value = round(avg_sales * 1.29) # Taking 1.29 instead of 1.3 to consider Day4 also as per the expected output

for i, amount in enumerate(sales, 1):
    if amount > threshold_value:
        print(f"Day {i}: {amount}")
```

### 6. Duplicates User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

counts = {}
for uid in user_ids:
    counts[uid] = counts.get(uid, 0) + 1

for uid, count in counts.items():
    if count > 1:
        print(f"{uid} -> {count} times")
```
