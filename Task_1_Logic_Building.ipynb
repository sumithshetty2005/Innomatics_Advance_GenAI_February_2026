### 1. User Login Check

### 2. Pass/Fail Analyzer

```python
marks = [45, 78, 90,33, 60]
successful_count = 0
unsuccessful_count = 0
for m in marks:
    if m >= 5:
        successful_count += 1
    else:
        unsuccessful_count += 1

print(f"Total Number of Pass Students: {successful_count}")
print(f"Total Number of Fail Students: {unsuccessful_count}")
```

### 3. Simple Data Cleaner

```python
names = ["ALice", "bob", "CHARLIE "]
cleaned_names_list = []

for name in names:
    cleaned_names_list.append(name.strip().lower())

print(cleaned_names_list)
```

### 4. Message Length Analyzer

```python
messages = ["HI", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg)
    status = " (Flagged: Message longer than 10 characters)" if length > 10 else ""
    print(f"Message: '{msg}' | Length: {length}{status}")
```

### 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print(f"Total ERROR entries: {error_count}")
```
